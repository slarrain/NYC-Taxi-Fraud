```python
methods = ['kmeans', 'dbscan', 'meanshift', 'boost']
```

Read in the data and give common column names

```python
dfs = {methods[0]:pd.read_csv('kmeans_driver_scores.csv', names=['hack','prob']),
       methods[1]:pd.read_csv('dbscan_driver_scores.csv', names=['hack','prob']),
       methods[2]:pd.read_csv('meanshift_driver_scores.csv', names=['hack','prob']),
       methods[3]:pd.read_csv('driver_score_clean_boosting_final.csv').drop('difference', axis=1).rename(columns={'hack_license':'hack','normal_prob':'prob'})}
                                 
```

```python
for key, df in dfs.iteritems():
    print key, len(df)
```

Plot the distribution of driver scores for each method

```python
for key, df in dfs.iteritems():
    plt.figure()
    df.prob.hist(bins=100)
    plt.title(key)
```

```python
# Make sure the same drivers are in each dataframe
hacks = set(dfs['kmeans'].hack).intersection(dfs['boost'].hack)
```

```python
for key, df in dfs.iteritems():
    ix = df.hack.apply(lambda x: x in hacks)
    dfs[key] = df.ix[ix,:].sort('hack').reset_index(drop=True)
```

Join the separate driver score files into a single dataframe

```python
i = 0
for key, df in dfs.iteritems():
    if i == 0:
        old = df
        i = 1
    else:
        scores = old.join(df, rsuffix='_'+key)
        old = scores
```

```python
scores
```

Drop unneeded columns

```python
scores = scores.drop(['hack_kmeans','hack_dbscan','hack_boost'], axis=1).rename(columns={'prob':'prob_meanshift'})
```

Create columns for percentile rankings

```python
for key, df in dfs.iteritems():
    scores['percentile_'+key] = pd.DataFrame(df.prob.argsort()).reset_index().set_index('prob').add(1.).div(len(df))
```

Store scores with percentiles for later

```python
scores.to_csv('driver_scores_and_percentiles.csv')
```

Compare methods

```python
fig, ax = plt.subplots(1, 2, figsize=(12,4))
pref = 'percentile_'

m1 = 'meanshift'
m2 = 'boost'

s1 = scores[pref+m1]
s2 = scores[pref+m2]
for p in np.linspace(0,1,100, endpoint=False):
    ix = s1>p
    psame = sum(s2.ix[ix]>p)/float(sum(ix))
    ax[0].plot(p, psame, 'k.')

for p in np.linspace(0,1,100, endpoint=False):
    ix = s2>p
    psame = sum(s1.ix[ix]>p)/float(sum(ix))
    ax[1].plot(p, psame, 'k.')
```

Results plots

```python
x = np.linspace(0.95,1,500, endpoint=False)
y = []
pref = 'percentile_'
m1 = 'kmeans'
m2 = 'boost'

s1 = scores[pref+m1]
s2 = scores[pref+m2]
for p in x:
    ix = s1>p
    y += [sum(s2.ix[ix]>p)/float(sum(ix))]
    #plt.plot(p, psame, 'k.')
plt.plot(x, y, 'k-')
plt.ylim(0,1.)
plt.title('Rate of Agreement - K-Means/Boosting')
plt.xlabel('Driver Fraud Score Percentile')
plt.savefig('agreement_'+m1+'-'+m2+'.png')
```

Output drivers above 99th percentile according to k-means and boosting

```python
pcut = 0.99
u = 'kmeans'
s = 'boost'
hf = scores.ix[(scores[pref+u]>pcut)&(scores[pref+s]>pcut),:]
```

```python
hf
```

```python
hf['avg_prob'] = hf.prob_kmeans.add(hf.prob_boost).div(2.)
```

```python
hf.sort('avg_prob')[['hack','prob_kmeans','percentile_kmeans','prob_boost','percentile_boost','avg_prob']].reset_index(drop=True).to_csv('high_fraud_hacks.csv')
```
